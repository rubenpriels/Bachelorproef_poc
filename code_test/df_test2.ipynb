```python
import pandas as pd
```

```python
df = pd.DataFrame({
    "locatie": ["loc1"]*12 + ["loc2"]*12 + ["loc3"]*12,
    "start": ["2024-01-01"]*6 + ["2025-01-01"]*6 + ["2024-01-01"]*6 + ["2025-01-01"]*6 + ["2024-01-01"]*6 + ["2025-01-01"]*6,
    "eind": [None]*36,
    "category": ["A", "B", "C", "D", "E", "F"]*6,
    "value": [10, 20, 30, 40, 50, 60, 15, 25, 35, 45, 55, 65]*3
})
```

```python
df_out = df.pivot(
    index=["locatie", "start", "eind"], 
    columns="category", 
    values="value"
).reset_index()

df_out.columns.name = None
```

```python
df2 = pd.DataFrame({
    "locatie": ["loc1", "loc2"],
    "eigenschap": ["x", "y"]
})
```

```python
df_joined = pd.merge(df_out, df2, on="locatie", how="left")
df_joined
```
