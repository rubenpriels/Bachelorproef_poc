```python
import pandas as pd
```

```python
df = pd.DataFrame({
    "locatie": ["locatie1", "locatie1", "locatie1", "locatie2", "locatie2", "locatie2", "locatie3", "locatie3"],
    "category": ["A", "B", "C", "A", "B", "C", "B", "C"],
    "value": [10, 20, 30, 40, 50, 60, 70, 80]
})
```

```python
df
```

```python
df_out = df.pivot(index="locatie", columns="category", values="value").reset_index()
df_out.columns.name = None
df_out
```

```python
df.duplicated(subset=["locatie", "category"]).any()
```

```python
df = pd.DataFrame({
    "locatie": ["loc1"]*12 + ["loc2"]*12 + ["loc3"]*12,
    "start": ["2024-01-01"]*6 + ["2025-01-01"]*6 + ["2024-01-01"]*6 + ["2025-01-01"]*6 + ["2024-01-01"]*6 + ["2025-01-01"]*6,
    "eind": ["2024-12-31"]*6 + ["2025-12-31"]*6 + ["2024-12-31"]*6 + ["2025-12-31"]*6 + ["2024-12-31"]*6 + ["2025-12-31"]*6,
    "category": ["A", "B", "C", "D", "E", "F"]*6,
    "value": [10, 20, 30, 40, 50, 60, 15, 25, 35, 45, 55, 65]*3
})
```

```python
df
```

```python
df_out = df.pivot(
    index=["locatie", "start", "eind"], 
    columns="category", 
    values="value"
).reset_index()

df_out.columns.name = None
```
